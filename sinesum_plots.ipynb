**Ploting the graph**

Now we will plot for $n = 1, 3, 5, 10, 30, 100$ and $t = \alpha T$, with $T = 2\pi$ and $\alpha = 0.01, 0.25, 0.49$.  and you will notice that a n increases it becomes more accurates approximation.

````python
n = np.array([1, 3, 5, 10, 30, 100]) 
figure = plt.figure(figsize=(16,6)) 
t = np.linspace(-np.pi,np.pi, num=100, endpoint=True) 
s = np.vectorize(se.sn)
arr1 = np.array([s(t, num) for num in n])
axes = plt.axes() 

i = 0
for s in arr1: 
    axes.plot(t, s, label="n = {}".format(n[i])) 
    i=i+1 

axes.legend()
plt.xlabel("t") 
plt.ylabel("S_n(t)") 
plt.show() 
````
